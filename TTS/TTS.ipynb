```python

import googletrans
from googletrans import Translator
from gtts import gTTS
from IPython.display import Audio, display

# Initialize the Google Translate API translator
translator = Translator()

src_text = input("Enter the text to translate: ")
src = 'en'
dest = 'te'

# Translate the text from English to Telugu
translation = translator.translate(src_text, src=src, dest=dest)

# Print the translated text
print(translation.text)
```

```python
def text_to_speech(text, lang, filename="output.mp3"):
    try:
        tts = gTTS(text=text, lang=lang)
        tts.save(filename)
        return filename
    except ValueError as e:
        print(f"Error: {e}")
        return None

# Convert the translated text to speech
audio_file = text_to_speech(translation.text, lang=dest)
```

```python
if audio_file:
    # Play the audio file
    display(Audio(audio_file, autoplay=True))
else:
    print("Failed to convert text to speech.")
```
